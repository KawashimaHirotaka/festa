## Import Libraries

```
import style_transfer as st
import matplotlib.pyplot as plt
```

## Download Style & Content images

対象の画像 と 適応したい画風を入手します。
[Google](https://www.google.co.jp/imghp) で画像を検索し、 `URL` を指定してください。

```
st.fetch_img("https://pytorch.org/tutorials/_static/img/neural-style/picasso.jpg")
st.fetch_img("https://pytorch.org/tutorials/_static/img/neural-style/dancing.jpg")
```

ダウンロードした画像の保存場所と名称を確認します。

```
st.picture_list()
```

## Load `Style` & `Content`
ダウンロードした２枚の画像をコンピュータに読み込みます。

```
style = st.load_img("images/picasso.jpg")
content = st.load_img("images/dancing.jpg")
```

読み込んだ画像をそれぞれ表示して確認しましょう。

```
plt.figure(figsize=(20, 20))

plt.subplot(2, 2, 1)
st.img_show(style, title="Style")


plt.subplot(2, 2, 2)
st.img_show(content, title="Content")
```

## Run `Neural Style Transfer`

Neural Style Transfer を実行します。

処理が終了すると画風が適応された画像が表示されます。

画像によりますが 5分程度で終了します。

```
st.run(style_img=style, content_img=content)
```
